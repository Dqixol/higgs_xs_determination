```python
import ROOT
%jsroot on
```

```python
df = ROOT.RDataFrame("mc20e_ggf_hyy", "mc20e_ggf_hyy.root")
```

```python
model = ROOT.RDF.TH1DModel("h", "h", 100, 0, 1000)
h =df.Define("higgs_pt", "Higgs_p4.Pt() / 1e3").Histo1D(model, "higgs_pt", 'weight_mc')
c = ROOT.TCanvas()
h.Draw()
c.Draw()
```
