```python
import ROOT
xs = 4.852E+01 * 1e3 # in fb, see https://twiki.cern.ch/twiki/bin/view/LHCPhysics/CERNYellowReportPageAt13TeV
# xs = 1.0 # for testing
```

```python
df = ROOT.RDataFrame("mc20e_ggf_hyy", "mc20e_ggf_hyy.root")
sum_of_weights_all = df.Sum("weight_mc")
counts_all = df.Count()
df = df.Filter("abs(Higgs_p4.Rapidity()) < 2.5")
sum_of_weights_300plus = df.Filter('Higgs_p4.Pt() > 300e3').Sum("weight_mc")
sum_of_weights_300_450 = df.Filter('Higgs_p4.Pt() > 300e3').Filter('Higgs_p4.Pt() < 450e3').Sum("weight_mc")
sum_of_weights_450_650 = df.Filter('Higgs_p4.Pt() > 450e3').Filter('Higgs_p4.Pt() < 650e3').Sum("weight_mc")
counts_300plus = df.Filter('Higgs_p4.Pt() > 300e3').Count()
counts_300_450 = df.Filter('Higgs_p4.Pt() > 300e3').Filter('Higgs_p4.Pt() < 450e3').Count()
counts_450_650 = df.Filter('Higgs_p4.Pt() > 450e3').Filter('Higgs_p4.Pt() < 650e3').Count()
sum_of_weights_650plus = df.Filter('Higgs_p4.Pt() > 650e3').Sum("weight_mc")
counts_650plus = df.Filter('Higgs_p4.Pt() > 650e3').Count()
futures = [sum_of_weights_all, counts_all, sum_of_weights_300plus, sum_of_weights_300_450, sum_of_weights_450_650, counts_300plus, counts_300_450, counts_450_650, sum_of_weights_650plus, counts_650plus]
ROOT.RDF.RunGraphs(futures)
```

```python
print('mc20 ggf_hyy:')
print(f"incl xs     = {xs:<10.3e} fb, counts = {counts_all.GetValue()}")
print(f"300+ GeV    = {sum_of_weights_300plus.GetValue()/sum_of_weights_all.GetValue()*xs:<10.3e} fb, counts = {counts_300plus.GetValue()}")
print(f"300-450 GeV = {sum_of_weights_300_450.GetValue()/sum_of_weights_all.GetValue()*xs:<10.3e} fb, counts = {counts_300_450.GetValue()}")
print(f"450-650 GeV = {sum_of_weights_450_650.GetValue()/sum_of_weights_all.GetValue()*xs:<10.3e} fb, counts = {counts_450_650.GetValue()}")
print(f"650+ GeV    = {sum_of_weights_650plus.GetValue()/sum_of_weights_all.GetValue()*xs:<10.3e} fb, counts = {counts_650plus.GetValue()}")
```
